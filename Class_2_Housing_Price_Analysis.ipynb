```python
import requests
import certifi
import io

url = "https://raw.githubusercontent.com/fuyaoqiuuu/schulich_data_science/main/Datasets/kc_house_data.csv"
resp = requests.get(url, verify=certifi.where())
resp.raise_for_status()

df = pd.read_csv(io.StringIO(resp.text))
```
