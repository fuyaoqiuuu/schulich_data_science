Assume arr is a NumPy array of numbers — for example, the income sample from the slides: arr = np.array([45, 55, 60, 50, 80, 62, 54, 58, 71, 48]). 

NumPy (imported as np) is a library that lets us do fast math on whole arrays of numbers at once, instead of writing loops.This block calculates measures of dispersion — numbers that tell us how spread out the data is.

 Central tendency (mean, median, mode) tells us where the middle is; dispersion tells us how far values stray from that middle.

```python
data = [45, 55, 60, 50, 80, 62, 54, 58, 71, 48]

print(type(data)) # using the "type" function, allows you to check the data type you are working with (numeric, text, boolean etc)


# Numpy has a various functions you can use to perform EDA, one of the advantages of Numpy is that you can use it on arrays. The way you transform lists to array is np.array
arr = np.array(data) 
arr
```

### Measures of Central Tendency

```python
# one of the important ceoncepts when working with python is the use of variables. The way you initiate a variable is "variable_name = value"


mean_val = np.mean(arr)
median_val = np.median(arr)


# The reason we use stats for mode is because Numpy doesn't have a native mode function, so that's why we use the scipy stats for mode
mode_result = stats.mode(arr, keepdims=False)  # returns ModeResult(mode, count)
```

### To see the output of the functions you ran, simply use the "print" function. The reason we do is print allows us to show ALL the results, if we simply called the  3 variables without print, only the last output will be displayed, try it

```python
print(mean_val)
print(median_val)
print(mode_result.mode)
```

### Measures of Dispersion

```python
data_range = np.max(arr) - np.min(arr)
variance = np.var(arr, ddof=1)        # ddof=1 -> sample variance
std_dev = np.std(arr, ddof=1)         # ddof=1 -> sample standard deviation
q1, q3 = np.percentile(arr, [25, 75])
iqr = q3 - q1

print(data_range)
print(variance)
print(std_dev)
print(q1)
print(q3)
print(iqr)
```

Range = largest value minus smallest value. np.max(arr) returns the biggest number in the array, np.min(arr) returns the smallest. We subtract them to get the total span. It's the simplest measure of spread, but also the most fragile — a single extreme value can blow it up.

```python
data_range = np.max(arr) - np.min(arr)
print(data_range)
```

Variance measures the average squared distance of each value from the mean. We square the distances so positive and negative deviations don't cancel each other out.


The ddof=1 argument is important. "ddof" stands for Delta Degrees of Freedom. By default NumPy divides by n (the number of values), which gives you the population variance — correct if your data is the entire population. With ddof=1 it divides by n - 1 instead, giving you the sample variance — the right choice when your data is a sample from a larger population (which is almost always the case in data science). Dividing by n - 1 corrects a built-in bias and gives a more accurate estimate.
Variance is in squared units (squared dollars, squared years, etc.), which is hard to interpret — that's why we usually report standard deviation instead.

```python
variance = np.var(arr, ddof=1)        # ddof=1 -> sample variance
print(variance)
```

Standard deviation is the square root of variance. Taking the square root brings the number back into the original units, which is why it's the dispersion measure most people actually quote. A standard deviation of 10 on income (thousands) means typical incomes fall roughly 10 thousand away from the mean.

```python
std_dev = np.std(arr, ddof=1)         # ddof=1 -> sample standard 
print(std_dev)
```

Quartiles split the data into four equal-sized chunks when sorted.

Q1 (the 25th percentile) is the value below which 25% of the data falls.
Q3 (the 75th percentile) is the value below which 75% of the data falls.
Q2 is the median (50th percentile) — we don't compute it here, but that's where it would sit.

np.percentile(arr, [25, 75]) asks for both percentiles in one call and returns them as a pair. We unpack that pair into two variables — q1 and q3 — in a single line. This is called tuple unpacking in Python and saves us from writing two separate calls.

```python
q1, q3 = np.percentile(arr, [25, 75])
```

IQR (Interquartile Range) = Q3 minus Q1. It's the spread of the middle 50% of the data.
Why care about the middle 50%? Because it ignores the extremes on both ends. Unlike the range, a single outlier can't distort the IQR. This is why the IQR is the spread measure of choice when your data has outliers or is skewed — exactly like income data. It's also what powers the box in a boxplot, and the basis for the 1.5×IQR rule for flagging outliers (which comes up later in the module).

```python
iqr = q3 - q1
print(iqr)
```

The method below is shows using an "f-string". An f-string (short for "formatted string literal") is a string that lets you drop variables and expressions directly inside it. You mark the string as an f-string by putting the letter f right before the opening quote.

```python
print(f"Range:    {data_range}")
print(f"Variance: {variance:.2f}")    # the :.2f rounds to 2 decimal places
print(f"Std Dev:  {std_dev:.2f}")
print(f"Q1:       {q1}")
print(f"Q3:       {q3}")
print(f"IQR:      {iqr}")
```

Examples to better understand an f-String

```python
name = "Alice"
print(f"Hello, {name}!")
```

The magic happens inside the { } curly braces. Python evaluates whatever is inside the braces and substitutes the result into the string.

Why use f-String?

Compare these three ways of doing the same thing:


```python
mean = 58.3

# Old way 1: string concatenation (clunky, needs str() conversion)
print("The mean is " + str(mean))

# Old way 2: .format() method (works, but verbose)
print("The mean is {}".format(mean))

# Modern way: f-string (clean, readable)
print(f"The mean is {mean}")
```
