```python
def has_duplicates_slow(nums):
    # Outer loop runs 'n' times
    for i in range(len(nums)):
        # Inner loop also runs up to 'n' times
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                return True # Found a duplicate!
        breakpoint()
    return False
```

```python
# additional memory (RAM) for this method
def has_duplicates_fast(nums):
    seen = set() # Extra memory to store unique numbers
    # This loop runs exactly 'n' times
    for num in nums:
        breakpoint()
        if num in seen: # Checking a set is O(1) instantly!
            return True
        seen.add(num)

    return False
```

### Using the Debugger with Breakpoints


```python
test_list = [1, 2, 3, 4, 5]
has_duplicates_fast(test_list)
```
