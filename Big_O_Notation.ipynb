```python
def has_duplicates_slow(nums):
    # Outer loop runs 'n' times
    for i in range(len(nums)):
        # Inner loop also runs up to 'n' times
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                return True # Found a duplicate!
    return False
```
