```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

print('Libraries loaded.')
print(f'pandas {pd.__version__  }, numpy {np.__version__}')

```

```python
data = [45,55,60,50,80,62,54,58,71,48]

arr = np.array(data)
arr 
```

```python
mean_val = np.mean(arr)
median_val = np.median(arr)
mode_val = stats.mode(arr, keepdims=False)

print(f'Mean: {mean_val:.2f} (sensitive to outliers)')
print(f'Median: {median_val:.2f} (robust to outliers)')
print(f'Mode: {mode_val.mode} (occurs {mode_val.count} times(s))')
```
